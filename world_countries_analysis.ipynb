```python
# A Pandas Portfolio Project 
# Step 1: Load & Preview the data
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv(r"C:\Users\sharm\Downloads\countries_and_territories.csv")
```

```python
# Preview first 5 rows
df.head()
```

```python
# Preview last 5 rows
df.tail()
```

```python
# Step 2: Understand the dataset
# Check shape
print(df.shape)
```

```python
# Column names
print(df.columns)
```

```python
# Data types
print(df.dtypes)
```

```python
# Basic statistics
print(df.describe)
```

```python
# Step 3: Clean the Data
# Check missing values
print(df.isnull().sum())
```

```python
# Check for duplicates
print(df.duplicated().sum())
```

```python
# Reset the index cleanly
df = df.reset_index(drop=True)
```

```python
# Quick sanity check
df.head()
```

```python
#  Step 4: Explore & Analyze
#  Top 10 most populous countries
top10_pop = df.nlargest(10, 'Population')
print(top10_pop[['Country','Population']])
```

```python
# Top 10 largest countries by area
top10_area = df.nlargest(10, 'Area')
print(top10_area[['Country','Area']])
```

```python
# Count countries per region
region_counts = df['Region'].value_counts()
print(region_counts)
```

```python
# Average population by region
avg_pop = df.groupby('Region')['Population'].mean()
print(avg_pop.sort_values(ascending=False))
```

```python
# Countries with literacy rate below 50%
low_literacy = df[df['Literacy'] < 50]
print(low_literacy[['Country', 'Literacy']])
```

```python
# Top 10 richest countries by GDP per capital
top10_gdp = df.nlargest(10, 'GDP_per_capita')
print(top10_gdp[['Country', 'GDP_per_capita']])
```

```python
# Average GDP per region
avg_gdp = df.groupby('Region')['GDP_per_capita'].mean()
print(avg_gdp.sort_values(ascending=False))
```

```python
# Step 5: Visualizations
# Bar Chart: Top 10 Most Populous Countries
plt.figure(figsize=(12, 6))
plt.bar(top10_pop['Country'], top10_pop['Population'], color='steelblue')
plt.title('Top 10 Most Populous Countries')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Bar Chart: Top 10 Largest Countries by Area
plt.figure(figsize=(12, 6))
plt.bar(top10_area['Country'], top10_area['Area'], color='coral')
plt.title('Top 10 Largest Countries by Area')
plt.xlabel('Country')
plt.ylabel('Area (sq km)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Countries by Region
plt.figure(figsize=(8, 8))
plt.pie(region_counts.values,
        labels=region_counts.index,
        autopct='%1.1f%%',
        startangle=140)
plt.title('Distribution of Countries by Region')
plt.tight_layout()
plt.show()
```

```python
#  Bar Chart: Average GDP by Region
plt.figure(figsize=(12, 6))
avg_gdp.sort_values().plot(kind='barh', color='mediumseagreen')
plt.title('Average GDP per Capita by Region')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
# Histogram: Literacy Rate Distribution
plt.figure(figsize=(10, 5))
plt.hist(df['Literacy'], bins=15, color='mediumpurple', edgecolor='white')
plt.title('Literacy Rate Distribution Across Countries')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
